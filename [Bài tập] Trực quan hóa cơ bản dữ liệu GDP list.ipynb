```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu
df = pd.read_csv("GDPList.csv", encoding = "ISO-8859-1")
df.head()
```

```python
# Thông tin bộ dữ liệu
df.info()
```

```python
# Mô tả bộ dữ liệu
df.describe()
```

##### So sánh GDP các nước ở South America.

```python
# Lấy dữ liệu
df_1 = df[df["Continent"] == "South America"]
df_1.head()
```

```python
# Vẽ biểu đồ
plt.bar(df_1["Country"], df_1["GDP (millions of US$)"], color = "red")
plt.title("So sánh GDP các nước ở South America", fontsize = 14)
plt.xlabel("Tên nước", fontsize = 14)
plt.ylabel("GDP", fontsize = 14)
plt.show()
```
