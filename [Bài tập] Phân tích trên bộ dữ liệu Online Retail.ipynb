- Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
# Đọc dữ liệu
df = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
df.head()
```

- Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia.

```python
df.pivot_table(values = "Quantity", index = "InvoiceNo", columns = "Country", aggfunc = "mean")
```

- Xây dựng bảng Pivot table, với mỗi khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo kho.

```python
df.pivot_table(values = "Quantity", index = "StockCode", columns = "CustomerID", aggfunc = {min, max})
```

- Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.

```python
import numpy as np
```

```python
df.pivot_table(values = ["Quantity", "UnitPrice"], index = "InvoiceNo", columns = "StockCode", aggfunc = {"Quantity": np.sum, "UnitPrice": np.mean})
```

- Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được mỗi ngày

```python
df.pivot_table(values = "Quantity", index = "InvoiceDate", aggfunc = {"Quantity": np.sum})
```
