```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu
df = pd.read_csv("shopeep_koreantop_clothing_shop_data.csv")
df.head()
```

```python
# Thông tin bộ dữ liệu
df.info()
```

```python
# Mô tả bộ dữ liệu
df.describe()
```

- Vẽ biểu đồ so sánh số lượng shop gia nhập theo các năm.

```python
df_1 = df.groupby(by = ["join_year"]).count()
```

```python
plt.bar(df_1.index, df_1["pk_shop"], width = 0.5, color = ["blue", "red", "yellow"])
plt.title("So sánh số lượng shop gia nhập theo các năm", fontsize = 16)
plt.xlabel("Năm", fontsize = 14)
plt.ylabel("Số lượng shop gia nhập", fontsize = 14)
plt.show()
```
