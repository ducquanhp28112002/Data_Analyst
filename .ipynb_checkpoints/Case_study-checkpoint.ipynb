##### Tạo lưới để hiển thị

Có rất nhiều cách để tạo layout

```python
def create_grid():
    fig = plt.figure(constrained_layout = True, figsize = (10, 6))
    # fig = plt.figure(figsize = (16, 6))
    specs = gridspec.GridSpec(ncols = 12, nrows = 2, figure = fig) ### Declaring 12x2 figure.
    row1_axes = []
    for i in range(3):
        row1_axes.append(fig.add_subplot(specs[0, i*4:i*4+4])) ## First Row
    row2_axes = []
    for i in range(4):
        row2_axes.append(fig.add_subplot(specs[1, i*3:i*3+3])) ## Second Row
    
    # Chèn đồ thì 1 vào ô thứ nhất
    ax = sns.countplot(x = "Education", data = data, ax = row1_axes[0])
    # 2.
    sns.countplot(x = "Marital_Status", data = data, ax = row1_axes[1])
    row1_axes[1].set_xticklabels(row1_axes[1].get_xticklabels(), rotation = 30)
    #3. 
    pie_lables = [s.replace("Num", "").replace("Purchases", "") for s in df_TP1.index]
    row1_axes[2].pie(df_TP,labels = pie_lables, explode = myexplode, shadow = True, autopct = "%1.2f%%")

    sns.barplot(x = data["Marital_Status"], y = data["NumStorePurchases"], ax = row2_axes[0])
  
    sns.barplot(x = data["Marital_Status"], y = data["NumCatalogPurchases"], ax = row2_axes[1])

    sns.barplot(x = data["Marital_Status"], y = data["NumWebPurchases"], ax = row2_axes[2])

    sns.barplot(x = data["Marital_Status"], y = data["NumDealsPurchases"], ax = row2_axes[3])

    for ax in row2_axes:
        ax.set_xticklabels(ax.get_xticklabels(), rotation = 45) 

    plt.close(fig)
    return fig
```

##### Vẽ lại các biểu đồ lọc dữ liệu theo min, max của income

- Cách 1 sửa code trong hàm create_grid. Nhược điểm là làm hàm này trở nên phức tạp.
- Cách là viết lại các hàm tạo sơ đồ. Ưu điểm, chương trình sáng sủa, dễ phát triển và maintain

###### Viết lại các hàm vẽ đồ thị

So sánh chi tiêu theo trình độ học vấn

```python
def draw_compare_by_education(ax, df, min_income, max_incom):
    data_pipline = data[(data["Income"] >= min_income) & (data["Income"] <= max_incom)]
    sns.countplot(x = "Education", data = data_pipline, ax = ax)
```

So sánh sức mua theo tình trang gia đình

```python
def draw_compare_by_marital_status(ax, df, min_income, max_incom):  
    data_pipline = data[(data["Income"] >= min_income) & (data["Income"] <= max_incom)]
    sns.countplot(x = "Marital_Status", data = data, ax = ax)
    ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)
```

Tỷ trọng bán hàng trên các kênh

```python
def draw_compare_by_channel(ax, df, min_income, max_incom):
  
    data_pipline = data[(data["Income"] >= min_income) & (data["Income"] <= max_incom)]

    total_purchase = [col for col in data_pipline if col.startswith("Num")]
    df_1 = data_pipline[total_purchase].sum()
    df_2  = pd.DataFrame(df_TP)

    pie_lables = [s.replace("Num", "").replace("Purchases", "") for s in df_2.index]
    ax.pie(df_1,labels = pie_lables, explode = myexplode, shadow = True, autopct = "%1.2f%%")
```

Các biểu đồ so sánh chi tiết chi tiêu theo tình trạng gia đình trên từng loại mặt hàng

```python
def draw_compare_by_marital_status_1(ax, df, min_income, max_incom, marital_status):
    data_pipline = data[(data["Income"] >= min_income) & (data["Income"] <= max_incom)]
    sns.barplot(x = data_pipline["Marital_Status"], y = data_pipline[marital_status], ax = ax)
    ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
```

```python
def create_config(min_incom, max_income):
    with plt.style.context(("seaborn", "ggplot")):
        # fig = plt.figure(constrained_layout = True, figsize = (16, 6))
        fig = plt.figure( figsize=(16,6))    
        specs = gridspec.GridSpec(ncols = 12, nrows = 2, figure = fig) ### Declaring 12x2 figure.
        row1_axes = []
        for i in range(3):
            row1_axes.append(fig.add_subplot(specs[0, i * 4: i * 4 + 4])) ### First Row
        row2_axes = [] 
        for i in range(4):
            row2_axes.append(fig.add_subplot(specs[1, i * 3: i * 3 + 3])) ## Second Row

        #1
        # sns.countplot(x = "Education", data = data, ax = row1_axes[0])
        draw_compare_by_education(row1_axes[0], data, min_incom, max_income)
        #2. 
        draw_compare_by_marital_status(row1_axes[1], data, min_incom, max_income)
        # plt.xticks(rotation = 57)
        # sns.countplot(x = "Marital_Status", data = data, ax = row1_axes[1])
        # row1_axes[1].set_xticklabels(row1_axes[1].get_xticklabels(), rotation = 15)

        #3.
        draw_compare_by_channel(row1_axes[2], data, min_incom, max_income)
        # row1_axes[2].pie(df_TP, labels = df_TP1. index, explode = myexplode, shadow = True, autopct = "%1.2f%%")
    
        ls = ["NumStorePurchases", "NumCatalogPurchases", "NumWebPurchases", "NumDealsPurchases"]

        for idx, marital_status in enumerate(ls):
            # print("idx, marital_status", idx, marital_status)
            draw_compare_by_marital_status_1(row2_axes[idx], data, min_incom, max_income, marital_status)

        # plt.show()
        plt.close(fig)
        return fig
```

```python
create_config(data["Income"].min(), data["Income"].max())
```

```python
income_min = int(data["Income"].min())
income_max = int(data["Income"].max())
print("income_min", income_min, "income_max", income_max)
```

```python
## Define Panel widgets
min_income_slider = pn.widgets.IntSlider(name = "Min Income slider", start = income_min, end = income_max, step = 5, value = income_min)
min_income_slider
```

```python
# # Define Panel widgets
max_income_slider = pn.widgets.IntSlider(name = "Max Income slider", start = income_min, end = income_max, step = 5, value = income_min)
max_income_slider
```

Mục tiêu là lấy giá trị của min slider và max slider truyền vào hàm create_config để thay đổi các đồ thị

```python
create_config(2000, 300000)
```

Tạo Dashboard 2 có hỗ trợ min, max income

```python
@pn.depends(min_income_slider.param.value,max_income_slider.param.value)
def create_dashboard_2(min_ic, max_ic):
    # print("min_income_slider.param.value", min_income_slider.param.value)
    # min_ic=min_income_slider.param.value
    # max_ic=max_income_slider.param.value
    print("min_ic", min_ic)
    min_ic = float(min_ic)
    print("min_ic", min_ic, max_ic)
    max_ic = float(max_ic)
    return create_config(min_ic, max_ic)
```

```python
heading = "<h3>Dashboard 2: Phần tích bán hàng</h3>\nDạng dashboard có thể thay đổi hình vẽ, đồ thị khi thay đổi min - max income!"
text = pnw.StaticText(value = heading, align = "center")
dashboard_2 = pn.Column(  
    text,
    # pn.Row(pn.widgets.IntSlider(name = "Min Income slider", start = income_min, end = income_max, step = 5, value = income_min), align = "center"),
    # pn.Row(pn.widgets.IntSlider(name = "Max Income slider", start = income_min, end = income_max, step = 5, value = income_min), align = "center"),
    pn.Row(min_income_slider, align = "center"),
    pn.Row(max_income_slider, align = "center"),
    pn.Row(create_dashboard_2)
    )
# dashboard_2
```
