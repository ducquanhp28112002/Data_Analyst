```python
import pandas as pd
import openpyxl
```

```python
# Đọc dữ liệu
df = pd.read_excel("house_price_dống-da.xlsx")
df.head()
```

```python
# Kích thước bộ dữ liệu
df.shape
```

```python
# Thông tin bộ dữ liệu
df.info()
```

```python
# Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên
df[(df["ward_name"] == "Phường Trung Liệt") | (df["ward_name"] == "Phường Khâm Thiên")]
```

```python
# Lọc các thông tin địa chỉ, giá, hướng nhà, hướng ban công của các bản ghi có giấy chứng nhận số đỏ và có 3 phòng ngủ trở lên
df[(df["land_certificate"] == "Sổ đỏ") | (df["bedroom"] >= 6.0).filter(["address", "price", "house_direction", "balcony_direction"])]
```
