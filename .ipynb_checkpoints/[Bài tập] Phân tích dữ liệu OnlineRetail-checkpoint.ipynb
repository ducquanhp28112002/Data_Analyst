```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
df = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
df.head()
```

```python
# Số dong, số cột
df.shape
```

```python
# Thông tin bộ dữ liệu
df.info()
```

```python
# Mô tả bộ dữ liệu
df.describe()
```

- Tạo cột mới tên "Previous"

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["Month"] = df["InvoiceDate"].dt.month
df.head()
```

```python
conditions = [(df["Month"] == 1) | (df["Month"] == 2) | (df["Month"] == 3), (df["Month"] == 4) | (df["Month"] == 5) | (df["Month"] == 6), (df["Month"] == 7) | (df["Month"] == 8) | (df["Month"] == 9), (df["Month"] == 10) | (df["Month"] == 11) | (df["Month"] == 12)]
choices = [1, 2, 3, 4]
df["Previous"] = np.select(conditions, choices)
df.head()
```

- Tạo cột mới tên "Amout"

```python
df["Amount"] = df["Quantity"] * df["UnitPrice"]
df.head()
```

- Tạo cột mới tên "Discount"

```python
conditions = [(df["Country"] == "United Kingdom") & (df["Previous"] == 4), (df["Country"] == "France")]
choices = ["0.1", "0.05"]
df["Discount"] = np.select(conditions, choices, default = "0")
df
```

- Tạo cột mới tên "Total"

```python
df["Discount"] = pd.to_numeric(df["Discount"])
```

```python
df["Total"] = df["Amount"] * df["Discount"]
df.head()
```
