```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
df.head()
```

- Vẽ biểu đồ đường thể hiện xu hướng thay đổi số lượng đơn hàng theo thời gian trong năm 2011.

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df_1 = df[["InvoiceNo", "InvoiceDate"]]
df_1 = df_1.drop_duplicates(subset = "InvoiceNo", keep = "first")
df_1 = df_1.set_index(["InvoiceDate"])
df_2 = df_1["2011"]
df_2 = df_2.reset_index()
df_3 = df_2.groupby(by = df_2["InvoiceDate"].dt.date).count()
```

```python
# Vẽ biểu đồ đường
x = df_3.index.get_level_values(0)
plt.plot(x, df_3["InvoiceDate"])
plt.title("Number of Invoices in 2011 (Daily)", fontsize = 16)
plt.xlabel("Date", fontsize = 14)
plt.ylabel("Invoices", fontsize = 16)
plt.show()
```

##### Nhận xét
- Số lượng đơn hàng ngày có sự thay đổi mạnh.
- Giữa tháng 10 và 11 có 1 ngày có số đơn tăng đột biến.
- Tuy nhiên, biểu đồ này không thể hiện đuoẹc việc thiếu dữ liệu trong 1 số ngày (do cách nhóm dữ liệu).

- Vẽ biểu đồ cột so sánh số lượng đơn hàng trong các tháng của năm 2011

```python
df_4 = df_2.groupby(by = df_2["InvoiceDate"].dt.month).count()
df_4
```

```python
# Vẽ biểu đồ cột
x = df_4.index.get_level_values(0)
plt.bar(x, df_4["InvoiceDate"])
plt.title("Number of Invoices in 2011 (monthly)", fontsize = 16)
plt.xlabel("Month", fontsize = 14)
plt.ylabel("Invoices", fontsize = 14)
plt.show()
```

##### Nhận xét
- 3 tháng 9, 10, 11 có số lượng đơn hàng tăng đáng kể, trong đó tháng 11 có số đơn hàng cao nhất.
- Tháng 12 có tổng số lượng đơn hàng thấp nhất.
