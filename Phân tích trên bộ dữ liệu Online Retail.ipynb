```python
# Trích xuất dữ liệu từ 2 cột Description và Quantity
tg_1 = df.loc[:, ["Description", "Quantity"]]
tg_1.head(10)
```

```python
# Trích xuất dữ liệu 1000 dòng đầu tiên
tg_2 = df.iloc[0: 1000]
tg_2.tail(3)
```

```python
# Trích xuất các bản ghi có số lượng từ 10 trở lên
tg_3 = df.loc[df["Quantity"] >= 10]
tg_3.head(5)
```

```python
# Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice 
tg_4 = df.loc[1000: 2000, ["Quantity", "InvoiceDate", "UnitPrice"]]
tg_4.head()
```

```python
# Trích xuất các bản ghi có số hóa đơn ‘536365’
tg_5 = df.loc[df["InvoiceNo"] == 536365]
tg_5.head()
# Kết quả là không có
```
